```python
import numpy as np
from util.getmetrics import getmetrics
from modeling.statmodeling.RandomWalk import RandomWalk
from modeling.statmodeling.MeanModel import MeanModel
from modeling.statmodeling.SimpleMovingAverage import SimpleMovingAverage
```

```python
MO = 'new_deaths'
horizon = 4
actual, forecasts = RandomWalk(file_name = 'datasets/covid/OWID_weekly.csv',
                                training_ratio = 0.5, horizon = horizon, main_output = MO,
                                normalization = False)
print("Test set metrics {}".format(MO))
total_mse, total_mae, total_smape = 0, 0, 0
for i in range(horizon):
    mse, mae, smape = getmetrics(np.expand_dims(actual[:,i], axis = -1), np.expand_dims(forecasts[:,i], axis = -1))
    total_mse += mse
    total_mae += mae
    total_smape += smape
    print('h:{}, N:{}, MSE:{}, MAE:{}, sMAPE:{}'.format(i, len(actual), mse, mae, smape))
avg_mse = total_mse / horizon
avg_mae = total_mae / horizon
avg_smape = total_smape / horizon
print('Average MSE: {}, Average MAE: {}, Average sMAPE: {}'.format(avg_mse, avg_mae, avg_smape))
```
